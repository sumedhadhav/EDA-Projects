```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country, on = 'Country Code', how = 'left')
```

```python
final_df.head(2)
```

```python
# To check data types
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
# Pie Chart - Top 3 countries that use Zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
#Observation: Zomato's maximums records are from India. Followed by USA and UK
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# Observation 
1. Rating between 4.5 - 4.9 --- The rating is excellent
2. Rating between 4.0 - 4.4 --- The rating is very good
3. Rating between 3.5 - 3.9 --- The rating is good
4. Rating between 2.5 - 3.4 --- The rating is average
5. Rating between 1.8 - 2.4 --- The rating is Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,palette=['white','red','orange','yellow','green','green'])
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue ='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

```python
Observations 
1. Not rated count is the highest
2. Maximum numbers of ratings are between 2.5 to 3.4
```

```python
# Count plot
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
